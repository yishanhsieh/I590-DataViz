```python
import pandas as pd

# Assuming your dataset is stored in a DataFrame called df
# Example DataFrame structure
df_test = pd.DataFrame({'year': [2021, 2021, 2021], 'country': ['USA', 'India', 'Brazil'], 'gdp': [60000, 2500, 12000]})

# Calculate the quantiles for the GDP column
gdp_25 = df_test['gdp'].quantile(0.25)
gdp_50 = df_test['gdp'].quantile(0.50)
gdp_75 = df_test['gdp'].quantile(0.75)

# Define a function to assign GDP level based on the quantiles
def gdp_level(gdp):
    if gdp > gdp_75:
        return 'high'
    elif gdp_50 < gdp <= gdp_75:
        return 'higher-middle'
    elif gdp_25 < gdp <= gdp_50:
        return 'lower-middle'
    else:
        return 'low'

# Apply the function to create the 'gdp level' column
df_test['gdp level'] = df_test['gdp'].apply(gdp_level)

# Display the updated DataFrame
print(df_test)
```

```python
import pandas as pd

gdp_25 = df['Log GDP Per Capita'].quantile(0.25)
gdp_50 = df['Log GDP Per Capita'].quantile(0.50)
gdp_75 = df['Log GDP Per Capita'].quantile(0.75)

def gdp_level(gdp):
    if gdp > gdp_75:
        return 'high'
    elif gdp_50 < gdp <= gdp_75:
        return 'higher-middle'
    elif gdp_25 < gdp <= gdp_50:
        return 'lower-middle'
    else:
        return 'low'

df['gdp_level'] = df['Log GDP Per Capita'].apply(gdp_level)       
    
```

```python
# Life Ladder, 2005~2017
fig, ax = plt.subplots(figsize=(10, 6))
sns.boxplot(data=df, x='Year', y='Life Ladder', fill=False)
sns.stripplot(data=df, x='Year', y='Life Ladder', hue='gdp_level')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xlabel('Year')
plt.ylabel('Life Ladder')
plt.show()
```

```python
# Life Ladder, 2005~2017
# high gdp

filter_high_gdp = df[df['gdp_level']=='high']

fig, ax = plt.subplots(figsize=(10, 6))
sns.boxplot(data=filter_high_gdp, x='Year', y='Life Ladder', fill=False)
sns.stripplot(data=filter_high_gdp, x='Year', y='Life Ladder', hue='gdp_level')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xlabel('Year')
plt.ylabel('Life Ladder')
plt.show()
```
