```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("./joined_HPMD.csv")
df.info()
```

```python
df_2017 = df[df['Year']==2017]
max_row = df_2017.loc[df_2017['Log GDP Per Capita'].idxmax()]
min_row = df_2017.loc[df_2017['Log GDP Per Capita'].idxmin()]
max_country = max_row['Country Name']
min_country = min_row['Country Name']
max_country, min_country
```

```python
disorders = [
    "Schizophrenia", "Bipolar disorders", "Eating disorders", 
    "Anxiety disorders", "Drug use disorders", 
    "Depression", "Alcohol use disorders", "Life Ladder"
]
```

```python
df_2017_maxGDP = df_2017[df_2017['Country Name'] == max_country]
df_2017_minGDP = df_2017[df_2017['Country Name'] == min_country]

df_2017_maxGDP_filtered = df_2017_maxGDP[disorders] #index = 652
df_2017_minGDP_filtered = df_2017_minGDP[disorders] #index = 1184
print(df_2017_maxGDP_filtered)

df_2017_maxGDP_list = df_2017_maxGDP_filtered.loc[652].tolist()
df_2017_minGDP_list = df_2017_minGDP_filtered.loc[1184].tolist()
print(df_2017_maxGDP_list)
print(df_2017_minGDP_list)
```

```python
x = np.arange(len(disorders)) 

width = 0.35
# Plotting
plt.figure(figsize=(10, 6))
plt.bar(x - width/2, df_2017_maxGDP_list, width, label='Luxembourg',  color='skyblue')
plt.bar(x + width/2, df_2017_minGDP_list, width, label='Venezuela', color='lightgreen')

# Adding labels and title
plt.xlabel('Mental Disorders', fontsize=12)
plt.ylabel('Diagnosis Rate (%)', fontsize=12)
plt.title('Diagnosis Rates of Mental Disorders in Two Countries (2017)', fontsize=14)
plt.xticks(x, disorders, rotation=45, ha='right')
plt.legend()

# Display the chart
plt.tight_layout()
plt.show()
```
